```python
import numpy as np
from skimage.io import imread
```

```python
img = imread(r"Z:\Shared363\Shared\Microscopes\high-throughput-openframe\test\test_rotated.tif")
img.shape
```

```python
frame0 = img[0]
frame1 = img[1]
```

```python
# get projections of the frames
proj0_0 = np.sum(frame0, axis=0)
proj1_0 = np.sum(frame1, axis=0)
```

```python
# plot the projections
import matplotlib.pyplot as plt
plt.plot(proj0_0, 'r-')
plt.plot(proj1_0, 'b--')

plt.show()
```

```python
# get projections of the frames
proj0_1 = np.sum(frame0, axis=1)
proj1_1 = np.sum(frame1, axis=1)
```

```python
# plot the projections
plt.plot(proj0_1, 'r-')
plt.plot(proj1_1, 'b--')
plt.show()
```
